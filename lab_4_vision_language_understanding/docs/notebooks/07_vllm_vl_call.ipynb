```python
from openai import OpenAI
import base64
```

## vLLM on Inference

```python
vllm_client = OpenAI(base_url="http://10.0.10.66:8123/v1/",
                      api_key="support_vectors")
```

```python
model_name = "Qwen/Qwen2.5-VL-72B-Instruct"
```

```python
IMAGE_PATH = "../../images/phys_img.jpeg"

# ---- helper ----
def image_to_base64(path: str) -> str:
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

image_b64 = image_to_base64(IMAGE_PATH)
```

```python
from IPython.display import Image, display

display(Image(filename=IMAGE_PATH))
```

```python
response = vllm_client.chat.completions.create(
    model=model_name,
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": '''I am a physics student and came across this notes.  
                Can you explain it?  Put this explanation as a tutorial in a latex document.  
                Respond only the latex output with no other prefix text'''},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{image_b64}"
                    },
                },
            ],
        }
    ],
    max_tokens=5000,
)

print(response.choices[0].message.content)
```

```python
from rich import print as rprint
rprint(response)
```

```python
update_rule_image = "../../images/update_rule.jpeg"

# ---- helper ----
def image_to_base64(path: str) -> str:
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

image_b64 = image_to_base64(update_rule_image)
```

```python
from IPython.display import Image, display

display(Image(filename=update_rule_image))
```

```python
from litellm import completion
api_base = "http://10.0.10.66:8123/v1/"
api_key = "support_vectors"
response = completion(
    model=f"openai/{model_name}",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": '''Can you explain this diagram and the associated equation?
                I would also like to have a pytorch code implementation of this.'''},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{image_b64}"
                    },
                },
            ],
        }
    ],
    max_tokens=20000,
    api_base=api_base,
    api_key=api_key,
)
```

```python
from IPython.display import display, Markdown
response_text = response["choices"][0]["message"]["content"]
display(Markdown(response_text))
```
